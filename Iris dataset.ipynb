Exploratory Data Analysis 

```python
#plotting for species count using a bar chart

ax= sns.countplot(x='Species', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

We can see that all 3 species of Iris has 50 count.

```python
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',hue='Species',data=df)
```

* Iris-Virginica has the largest Petal length and width.
* Iris-Versicolor has the Medium Petal length and width.
* Iris-Setosa has the smaller Petal length and width compared to the other Species.


```python
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',hue='Species',data=df)
```

* Iris-Virginica has the largest Sepal length and smaller sepal width.
* Iris-Versicolor has the Medium Sepal length and width.
* Iris-Setosa has the smaller Sepal length and larger Sepal widths.
